```python
import os
import pandas as pd
```

```python
cwd = os.getcwd()
BASE_DIR = os.path.dirname(cwd)
DATA_DIR = os.path.join(BASE_DIR, 'data')
CACHE_DIR = os.path.join(BASE_DIR, 'cache')
os.makedirs(CACHE_DIR, exist_ok = True)
```

```python
for filename in os.listdir(DATA_DIR):
    print(filename)
```

```python
df = pd.read_csv(os.path.join(DATA_DIR, '2020_boxofficemojo_rankings.csv'))
df.head()
```

```python
df['Year'] = 2020
```

```python
df.head()
```

```python
my_entire_data = pd.DataFrame

my_dataframes = []
csv_files = [x for x in os.listdir(DATA_DIR) if x.endswith(".csv")]

for filename in csv_files:
    print(filename)
    year = filename.replace("_boxofficemojo_rankings.csv", "")
    csv_path = os.path.join(DATA_DIR, filename)
    this_df = pd.read_csv(csv_path)
    this_df['filename'] = filename
    this_df['year'] = year
    print(this_df.head(n=1))
    my_dataframes.append(this_df)
```

```python
my_entire_dataframe = pd.concat(my_dataframes)
```

```python
my_entire_dataframe.head()
```

```python
my_entire_dataframe.tail()
```

```python
dataset = os.path.join(CACHE_DIR, 'box-office-ranking-dataset.csv')
my_entire_dataframe.to_csv(dataset, index=False)
```
